```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import cv2 as cv
import glob
from sklearn import tree
import sklearn
import time
from sklearn.feature_selection import SelectKBest
import pygraphviz
from sklearn import tree
from sklearn import metrics
from sklearn.metrics import multilabel_confusion_matrix
import seaborn as sns
from sklearn.model_selection import GridSearchCV
```

```python
path = glob.glob("data/train/person/*.png")
cv_img = []
for img in path:
    n = np.array(cv.imread(img).flatten())
    cv_img.append(n)
```

```python
path = glob.glob("data/train/car/*.png")
for img in path:
    n = np.array(cv.imread(img).flatten())
    cv_img.append(n)
```

```python
path = glob.glob("data/train/dog/*.png")
for img in path:
    n = np.array(cv.imread(img).flatten())
    cv_img.append(n)
```

```python
path = glob.glob("data/train/airplane/*.png")
for img in path:
    n = np.array(cv.imread(img).flatten())
    cv_img.append(n)
```

```python
images = np.array(cv_img)
```

```python
images
```

```python
images.shape
```

```python
x_train = images
```

```python
y_ones = np.ones(500)
```

```python
y_zeros = np.zeros(1500)
```

```python
y_zeros.shape
```

```python
y_train = np.hstack((y_ones, y_zeros))
```

```python
y_train
```

```python
path = glob.glob("data/validation/person/*.png")
cv_img = []
for img in path:
    n = np.array(cv.imread(img).flatten())
    cv_img.append(n)
path = glob.glob("data/validation/car/*.png")
for img in path:
    n = np.array(cv.imread(img).flatten())
    cv_img.append(n)
path = glob.glob("data/validation/dog/*.png")
for img in path:
    n = np.array(cv.imread(img).flatten())
    cv_img.append(n)
path = glob.glob("data/validation/airplane/*.png")
for img in path:
    n = np.array(cv.imread(img).flatten())
    cv_img.append(n)
x_validation = np.array(cv_img)
y_validation = np.hstack((np.ones(100), np.zeros(300)))
```

```python
x_validation.shape
```

```python
y_validation.shape
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
start = time.time()
clf = clf.fit(x_train, y_train)
end = time.time()
print(f"Time taken to train is: {end - start}s")
```

```python
y_validation_predicted = clf.predict(x_validation)
```

```python
y_train_predicted = clf.predict(x_train)
```

```python
def accuracy(y_actual, y_pred):
    count = 0
    for i in range (len(y_pred)):
        if y_pred[i] == y_actual[i]:
            count += 1
    return (count*100/len(y_pred))
```

```python
print(sklearn.metrics.accuracy_score(y_train, y_train_predicted)*100)
print(sklearn.metrics.precision_score(y_train, y_train_predicted)*100)
print(sklearn.metrics.recall_score(y_train, y_train_predicted)*100)
```

```python
print(sklearn.metrics.accuracy_score(y_validation, y_validation_predicted)*100)
print(sklearn.metrics.precision_score(y_validation, y_validation_predicted)*100)
print(sklearn.metrics.recall_score(y_validation, y_validation_predicted)*100)
```

```python
# cm = metrics.confusion_matrix(y_validation, y_validation_predicted)
cm = metrics.confusion_matrix(y_train, y_train_predicted)
cm_df = pd.DataFrame(cm,
             index = [0,1], 
             columns = [0,1])

#Plotting the confusion matrix
plt.figure(figsize=(5,4))
sns.heatmap(cm_df, annot=True)
plt.title('Confusion Matrix')
plt.ylabel('Actual Values')
plt.xlabel('Predicted Values')
plt.show()
```

```python
selection = SelectKBest(k=10).fit(x_train, y_train)
```

```python
x_train_new = selection.transform(x_train)
```

```python
x_val_new = selection.transform(x_validation)
```

```python
x_train_new.shape
```

```python
x_val_new.shape
```

```python
clf = tree.DecisionTreeClassifier(max_depth = 10, min_samples_split = 7)
clf = clf.fit(x_train_new, y_train)
end = time.time()
print(f"Time taken to train is: {end - start}s")
```

```python
y_validation_predicted = clf.predict(x_val_new)
y_train_predicted = clf.predict(x_train_new)
```

```python
print(sklearn.metrics.accuracy_score(y_train, y_train_predicted)*100)
print(sklearn.metrics.precision_score(y_train, y_train_predicted)*100)
print(sklearn.metrics.recall_score(y_train, y_train_predicted)*100)
```

```python
print(sklearn.metrics.accuracy_score(y_validation, y_validation_predicted)*100)
print(sklearn.metrics.precision_score(y_validation, y_validation_predicted)*100)
print(sklearn.metrics.recall_score(y_validation, y_validation_predicted)*100)
```

```python
cm = metrics.confusion_matrix(y_validation, y_validation_predicted)
cm_df = pd.DataFrame(cm,
             index = [0,1], 
             columns = [0,1])

#Plotting the confusion matrix
plt.figure(figsize=(5,4))
sns.heatmap(cm_df, annot=True)
plt.title('Confusion Matrix')
plt.ylabel('Actual Values')
plt.xlabel('Predicted Values')
plt.show()
```
